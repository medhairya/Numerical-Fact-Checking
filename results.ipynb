```python
import json
from sklearn.metrics import classification_report, confusion_matrix

# ============================================
# 📌 Load Data
# ============================================
golden_path = r"E:\Projects\Numerical Fact Checking\CheckThat 2025\task3\data\1-my dataset\3-test-set-lables.json"
predicted_path = r"E:\Projects\Numerical Fact Checking\CheckThat 2025\task3\data\1-my dataset\3-veracity-done.json"

def load_json(path):
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)

golden_data = load_json(golden_path)
predicted_data = load_json(predicted_path)

print(f"Golden: {len(golden_data)} | Predicted: {len(predicted_data)}")

# ============================================
# 📌 Convert to {id: label}
# ============================================
gold = {x["id"]: x["label"] for x in golden_data}
pred = {x["id"]: x["label"] for x in predicted_data}

# ============================================
# 📌 Compare IDs
# ============================================
common_ids = gold.keys() & pred.keys()

missing_pred = gold.keys() - pred.keys()
missing_gold = pred.keys() - gold.keys()

print(f"Matched: {len(common_ids)}")
print(f"Missing in pred: {len(missing_pred)}")
print(f"Missing in gold: {len(missing_gold)}")

# ============================================
# 📌 Prepare labels
# ============================================
y_true = [gold[i] for i in common_ids]
y_pred = [pred[i] for i in common_ids]

labels = ["True", "Conflicting", "False"]

# ============================================
# 📌 Evaluation
# ============================================
print("\n📊 Classification Report:\n")
print(classification_report(y_true, y_pred, labels=labels, digits=4))

print("\n📊 Confusion Matrix:\n")
print(confusion_matrix(y_true, y_pred, labels=labels))

accuracy = sum(t == p for t, p in zip(y_true, y_pred)) / len(y_true)
print(f"\n✅ Accuracy: {accuracy:.4f}")
```
