# testing Linq-Embed-Mistral-FULL


```python
import torch
from transformers import AutoTokenizer, AutoModel

# 🔹 Your local model path
MODEL_PATH = r"E:\Embedding Models\Linq-Embed-Mistral-FULL"

# 🔹 Load tokenizer
print("Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH, trust_remote_code=True)

# 🔹 Load model (SAFE MODE)
print("Loading model...")
model = AutoModel.from_pretrained(
    MODEL_PATH,
    trust_remote_code=True
)
model.eval()

# 🔹 Sample text
texts = [
    "Stock market is bullish today",
    "Nifty is showing resistance near 22500"
]

# 🔹 Tokenize
inputs = tokenizer(
    texts,
    padding=True,
    truncation=True,
    return_tensors="pt"
)

# 🔹 Generate embeddings
with torch.no_grad():
    outputs = model(**inputs)

# 🔹 Mean pooling (IMPORTANT for embeddings)
embeddings = outputs.last_hidden_state.mean(dim=1)

# 🔹 Print results
print("\n✅ Embeddings shape:", embeddings.shape)
print("✅ First embedding sample:", embeddings[0][:10])  # first 10 values
```
