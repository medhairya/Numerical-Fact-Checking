# finding Unique docs for each claim and sort them using scores

```python
import json
from pathlib import Path

input_path = Path(r"E:\Projects\Numerical Fact Checking\CheckThat 2025\task3\data\1-my dataset\3-reranked.json")
output_path = Path(r"E:\Projects\Numerical Fact Checking\CheckThat 2025\task3\data\1-my dataset\3-processed-reranked.json")

with input_path.open("r", encoding="utf-8") as f:
    data = json.load(f)

processed_data = []

for item in data:
    claim = item.get("claim")
    original_id = item.get("original_id")
    docs = item.get("docs", [])
    scores = item.get("scores", [])

    # Keep max score for each unique doc
    doc_to_score = {}
    for doc, score in zip(docs, scores):
        if doc in doc_to_score:
            doc_to_score[doc] = max(doc_to_score[doc], score)
        else:
            doc_to_score[doc] = score

    # Sort unique docs by score descending
    sorted_docs_scores = sorted(doc_to_score.items(), key=lambda x: x[1], reverse=True)

    # Keep up to top 5
    top_docs_scores = sorted_docs_scores[:5]

    processed_item = {
        "claim": claim,
        "original_id": original_id,
        "number_of_unique_docs": len(doc_to_score),
        "docs": [doc for doc, score in top_docs_scores],
        "scores": [score for doc, score in top_docs_scores],
    }

    processed_data.append(processed_item)

with output_path.open("w", encoding="utf-8") as f:
    json.dump(processed_data, f, ensure_ascii=False, indent=2)

print(f"Saved {len(processed_data)} processed claims to:")
print(output_path)
```
